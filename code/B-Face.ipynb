```python
# 8-polyline-3D
# 2021-12-11
#
import cadquery as cq
import numpy as np
import pprint

from jupyter_cadquery.cadquery import show
from jupyter_cadquery import set_defaults, set_sidecar
#
# Comment/Uncomment  set_defaults() function
# if JupyterLab Theme ---> Dark  ; Uncomment it
# if JupyterLab Theme ---> Light ; use # make next line being a comment
set_defaults(axes=True, axes0=True, grid=True, ortho=False)
set_defaults(theme="dark") 
#
def wire_3D(points_3D_list) :
    my_wire_3D = (cq.Workplane().polyline(points_3D_list).close()
                    .wire()
                 )
    return my_wire_3D
# end def
#
points_3D_list = [[2*np.pi, 0, 0], [2*np.pi, 5, 0], [0, 5, 0], [0, 0, 0]]
motif = wire_3D(points_3D_list)
#
show(motif)
#
```
